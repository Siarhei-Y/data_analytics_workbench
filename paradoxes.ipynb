```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Генерация данных
np.random.seed(42)
x = np.linspace(0, 10, 50)
# Создаем данные для двух групп
group_a = pd.DataFrame({
    "group": "A",
    "x": x,
    "y": 8 * x + 2 + np.random.normal(scale=1.5, size=len(x)) 
})

x = np.linspace(5, 15, 50)
group_b = pd.DataFrame({
    "group": "B",
    "x": x,
    "y": 16 * x + 12 + np.random.normal(scale=1, size=len(x)) 
})

data = pd.concat([group_a, group_b])
```

```python
data
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Визуализация данных
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='x', y='y', color='red', linewidth=2,hue='group')
```
