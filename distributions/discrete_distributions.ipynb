```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss
```

# Дискретные распределения 

## Биномиальное распределение

Часто используемое распределение в тестах метрик сигналов (1,0). Независимая последовательность тестов Бернулли. Т.е. Дискретная случайная величина имеет **распределение Бернулли** c параметром $p∈[0, 1]$, если она принимает только значения 1 и 0 с вероятностями p и 1−p соответственно.
$$P(X=m)=C^m_np^nq^{n-m}$$
С параметрами n и p

мат ожидание: $M(x)=np$ <br>
дисперсия: $D(x)=npq$

```python
#симуляция выборки из биномиального распределения.
n = 10
p = 0.5
k = 10000 # сколько испытаний 
X = ss.binom(n=n,p=p)
X_vals = X.rvs(k)
plt.hist(X_vals,bins=100)

plt.title(f'Эмпирическая функция вероятности', fontsize=14)
plt.ylabel('Количество испытаний (k)', fontsize=12)
plt.xlabel(f'Кол-во успехов из {n} повторений в k испытаний', fontsize=12);
```


Чаще аналитикам приходится работать с частостями, то есть с самими вероятностями наступления события.
мат ожидание частости:$$M(\frac{n}{m})=p$$
дисперсия частости: $$D(\frac{n}{m})=\frac{pq}{n}$$

Стоит отметить, что $\mu=M(\frac{n}{m})$ и $\sigma=D(\frac{n}{m})$


```python
X_freq = X_vals / n
print('mu: ', round(np.mean(X_freq),2))
print('Мат.ожидание: ', p)
print('sigma: ',round(np.std(X_freq),2))
print('Дисперсия бином. распределения: ', round(np.sqrt(0.5**2/n),2))
```

```python
plt.hist(X_freq, bins=100)
plt.title(f'Эмпирическое распределение частостей', fontsize=14)
plt.ylabel('Количество испытаний (k)', fontsize=12)
plt.xlabel('Эмпирическая вероятность частот', fontsize=12);
```

##  Геометрическое расределение

Геометрическое распределение дискретной величины моделирует вероятность количества попыток до "успеха" . Т.е. нас волнут номер первого успеха, после которого эксперимент останавливается.

Дискретная случайная величина имеет **геометрическое распределение** с параметром $p∈[0, 1]$, если функция распределения вероятностей задана как:

$$
p_X(k) = P(X = k) = 
\begin{cases} 
(1 - p)^{k - 1}p, & \text{для } k \geq 1 \\
0, & \text{иначе}
\end{cases}
$$

Пример:
- Бросаем два кубика, интересуемся, когда выпадет 12.

```python
#симуляция выборки из равномерного распределения.
# n = 10
p = 2/12
n_size = 10000 # сколько раз 
X = ss.geom(p=p)
X_vals = X.rvs(n_size)

plt.figure(figsize=(10, 6))
plt.hist(X_vals,bins=100)
plt.title(f'Эмпирическое Геометрическое распределение (p = {round(p,3)})', fontsize=14);
```

```python
x = np.arange(1,20,1)
y = X.pmf(x)
```

```python
plt.bar(x,y)
plt.xticks(x)

plt.title(f'Геометрическое распределение (p = {round(p,3)})', fontsize=14)
plt.xlabel('Количество испытаний (k)', fontsize=12)
plt.ylabel('Вероятность P(X=k)', fontsize=12);
```
